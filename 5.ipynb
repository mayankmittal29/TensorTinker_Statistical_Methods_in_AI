## 5) AutoEncoder

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import precision_recall_fscore_support, roc_curve, auc, precision_recall_curve
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.patches import Patch
import pandas as pd
# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# Define hyperparameters
batch_size = 128
epochs = 40
learning_rate = 1e-3
bottleneck_dimensions = [8, 16, 32]  # Different bottleneck dimensions to compare
normal_digit = 4  # Change this to the last digit of your roll number

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Data loading and preprocessing
transform = transforms.Compose([
    transforms.ToTensor(),
])

# Load MNIST dataset
train_dataset = datasets.MNIST(root='./data', train=True, transform=transform, download=True)
test_dataset = datasets.MNIST(root='./data', train=False, transform=transform, download=True)

# Filter training dataset to only include normal digits
normal_indices = torch.tensor(train_dataset.targets) == normal_digit
normal_train_dataset = Subset(train_dataset, np.where(normal_indices)[0])

# Create data loaders
normal_train_loader = DataLoader(normal_train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Define the Autoencoder model
class Autoencoder(nn.Module):
    def __init__(self, bottleneck_dim):
        super(Autoencoder, self).__init__()
        
        # Encoder
        self.encoder = nn.Sequential(
            nn.Linear(28*28, 256),
            nn.ReLU(),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, bottleneck_dim),
            nn.ReLU()
        )
        
        # Decoder
        self.decoder = nn.Sequential(
            nn.Linear(bottleneck_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 256),
            nn.ReLU(),
            nn.Linear(256, 28*28),
            nn.Sigmoid()  # Output values between 0 and 1
        )
    
    def forward(self, x):
        # Encode
        encoded = self.encoder(x)
        # Decode
        decoded = self.decoder(encoded)
        return decoded

# Function to visualize original and reconstructed images
def visualize_reconstruction(model, test_loader, num_images=10):
    model.eval()
    images, labels = next(iter(test_loader))
    
    # Select a mix of normal and anomalous samples
    selected_indices = []
    normal_count = 0
    anomaly_count = 0
    
    for i, label in enumerate(labels):
        if label == normal_digit and normal_count < num_images // 2:
            selected_indices.append(i)
            normal_count += 1
        elif label != normal_digit and anomaly_count < num_images // 2:
            selected_indices.append(i)
            anomaly_count += 1
        
        if normal_count + anomaly_count >= num_images:
            break
    
    images = images[selected_indices]
    labels = labels[selected_indices]
    
    with torch.no_grad():
        images_flat = images.view(images.size(0), -1).to(device)
        reconstructions = model(images_flat).cpu().view(images.size(0), 1, 28, 28)
    
    # Calculate reconstruction errors
    reconstruction_errors = torch.sum((reconstructions.view(images.size(0), -1) - 
                                      images.view(images.size(0), -1)) ** 2, dim=1).numpy()
    
    fig, axes = plt.subplots(nrows=2, ncols=num_images, figsize=(num_images*2, 4))
    
    for i in range(num_images):
        # Original image
        axes[0, i].imshow(images[i].squeeze().numpy(), cmap='gray')
        axes[0, i].set_title(f"Original: {labels[i].item()}")
        axes[0, i].axis('off')
        
        # Reconstructed image
        axes[1, i].imshow(reconstructions[i].squeeze().numpy(), cmap='gray')
        is_normal = "Normal" if labels[i] == normal_digit else "Anomaly"
        axes[1, i].set_title(f"Recon: {is_normal}\nError: {reconstruction_errors[i]:.2f}")
        axes[1, i].axis('off')
        
    plt.tight_layout()
    return fig

# Function to train the model
def train_model(model, train_loader, num_epochs, learning_rate):
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    
    training_losses = []
    
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        
        for data in train_loader:
            img, _ = data
            img = img.view(img.size(0), -1).to(device)
            
            # Forward pass
            output = model(img)
            loss = criterion(output, img)
            
            # Backward and optimize
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item()
            
        epoch_loss = running_loss/len(train_loader)
        training_losses.append(epoch_loss)
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.6f}')
    
    # Plot training loss curve
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, num_epochs+1), training_losses, marker='o')
    plt.title(f'Training Loss for Bottleneck Dimension {model.encoder[-2].out_features}')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.grid(True)
    plt.savefig(f'training_loss_dim_{model.encoder[-2].out_features}.png')
    plt.show()
    
    return model, training_losses

# Function to compute reconstruction errors
def compute_reconstruction_error(model, data_loader):
    model.eval()
    reconstruction_errors = []
    labels = []
    
    with torch.no_grad():
        for data in data_loader:
            img, label = data
            img = img.view(img.size(0), -1).to(device)
            
            # Forward pass
            output = model(img)
            
            # Compute reconstruction error for each sample in the batch
            error = torch.sum((output - img) ** 2, dim=1).cpu().numpy()
            reconstruction_errors.extend(error)
            labels.extend(label.numpy())
    
    return np.array(reconstruction_errors), np.array(labels)

# Function to evaluate performance with different metrics
def evaluate_performance(reconstruction_errors, labels, threshold=None):
    # Convert labels to binary (normal=1, anomaly=0)
    binary_labels = (labels == normal_digit).astype(int)
    
    # If threshold is not provided, find the optimal threshold using PR curve
    if threshold is None:
        precision, recall, thresholds = precision_recall_curve(binary_labels, -reconstruction_errors)
        f1_scores = 2 * (precision * recall) / (precision + recall + 1e-10)
        optimal_idx = np.argmax(f1_scores)
        threshold = thresholds[optimal_idx] if optimal_idx < len(thresholds) else thresholds[-1]
    
    # Predict based on reconstruction error and threshold
    predictions = (reconstruction_errors <= -threshold).astype(int)
    print("Binary labels distribution:", np.sum(binary_labels), "out of", len(binary_labels))
    print("Predictions distribution:", np.sum(predictions), "out of", len(predictions))
    # Calculate metrics
    precision, recall, f1, _ = precision_recall_fscore_support(binary_labels, predictions, average='binary')
    
    # Calculate ROC curve and AUC
    fpr, tpr, _ = roc_curve(binary_labels, -reconstruction_errors)
    roc_auc = auc(fpr, tpr)
    
    return {
        'threshold': threshold,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'auc': roc_auc,
        'fpr': fpr,
        'tpr': tpr
    }

# Function to plot histograms of reconstruction errors with fun hues
def plot_error_histogram(reconstruction_errors, labels, threshold, bottleneck_dim):
    plt.figure(figsize=(12, 7))
    
    # Create a DataFrame for seaborn
    df = pd.DataFrame({
        'Reconstruction Error': reconstruction_errors,
        'Type': ['Normal' if label == normal_digit else 'Anomaly' for label in labels],
        'Digit': labels
    })
    
    # Create custom color palette
    normal_color = "#3498db"  # Blue
    anomaly_color = "#e74c3c"  # Red
    
    # Plot using seaborn with improved visuals
    ax = sns.histplot(
        data=df, 
        x='Reconstruction Error', 
        hue='Type',
        palette={
            'Normal': normal_color,
            'Anomaly': anomaly_color
        },
        kde=True,
        element="step",
        bins=50,
        alpha=0.6
    )
    
    # Add threshold line
    plt.axvline(x=-threshold, color='#2ecc71', linestyle='--', linewidth=2, 
                label=f'Threshold: {-threshold:.4f}')
    
    # Enhance visualization
    plt.title(f'Distribution of Reconstruction Errors (Bottleneck Dim: {bottleneck_dim})', fontsize=16)
    plt.xlabel('Reconstruction Error', fontsize=14)
    plt.ylabel('Count', fontsize=14)
    plt.legend(title='', fontsize=12)
    
    # Add annotation for accuracy
    normal_errors = reconstruction_errors[labels == normal_digit]
    anomaly_errors = reconstruction_errors[labels != normal_digit]
    
    true_positives = sum(normal_errors <= threshold)
    true_negatives = sum(anomaly_errors > threshold)
    
    accuracy = (true_positives + true_negatives) / len(reconstruction_errors)
    
    plt.annotate(f'Accuracy: {accuracy:.4f}', 
                 xy=(0.75, 0.9), 
                 xycoords='axes fraction',
                 bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))
    
    plt.tight_layout()
    plt.savefig(f'error_histogram_dim_{bottleneck_dim}.png', dpi=300)
    plt.show()
    
    return accuracy

# Function to create heatmap of reconstruction errors by digit
def plot_error_heatmap_by_digit(reconstruction_errors, labels, bottleneck_dim):
    plt.figure(figsize=(10, 6))
    
    # Calculate average reconstruction error for each digit
    error_by_digit = {}
    for digit in range(10):
        digit_errors = reconstruction_errors[labels == digit]
        if len(digit_errors) > 0:
            error_by_digit[digit] = np.mean(digit_errors)
    
    # Create heatmap data
    digits = list(range(10))
    errors = [error_by_digit.get(d, 0) for d in digits]
    
    # Create colormap: green for normal, red for anomaly
    colors = ['#e74c3c' if d != normal_digit else '#2ecc71' for d in digits]
    
    # Plot
    bars = plt.bar(digits, errors, color=colors)
    plt.title(f'Average Reconstruction Error by Digit (Bottleneck Dim: {bottleneck_dim})', fontsize=14)
    plt.xlabel('Digit', fontsize=12)
    plt.ylabel('Average Reconstruction Error', fontsize=12)
    plt.xticks(digits)
    
    # Add labels
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                f'{height:.2f}', ha='center', va='bottom', fontsize=10)
    
    # Add normal/anomaly legend
    legend_elements = [
        Patch(facecolor='#2ecc71', label=f'Normal (Digit {normal_digit})'),
        Patch(facecolor='#e74c3c', label='Anomaly (Other Digits)')
    ]
    plt.legend(handles=legend_elements)
    
    plt.tight_layout()
    plt.savefig(f'error_by_digit_dim_{bottleneck_dim}.png', dpi=300)
    plt.show()

# Function to plot precision-recall curve
def plot_precision_recall_curve(reconstruction_errors, labels, bottleneck_dim):
    # Convert labels to binary (normal=1, anomaly=0)
    binary_labels = (labels == normal_digit).astype(int)
    
    # Calculate precision-recall curve
    precision, recall, thresholds = precision_recall_curve(binary_labels, -reconstruction_errors)
    
    # Calculate F1 scores for each threshold
    f1_scores = 2 * (precision * recall) / (precision + recall + 1e-10)
    
    # Find optimal threshold
    optimal_idx = np.argmax(f1_scores)
    optimal_threshold = thresholds[optimal_idx] if optimal_idx < len(thresholds) else thresholds[-1]
    
    # Calculate average precision
    ap = np.sum((recall[:-1] - recall[1:]) * precision[:-1])
    
    # Plot precision-recall curve
    plt.figure(figsize=(10, 6))
    plt.plot(recall, precision, color='#3498db', linewidth=2)
    plt.scatter(recall[optimal_idx], precision[optimal_idx], s=100, c='#e74c3c', 
               label=f'Optimal Threshold: {optimal_threshold:.4f}')
    
    plt.title(f'Precision-Recall Curve (Bottleneck Dim: {bottleneck_dim})', fontsize=14)
    plt.xlabel('Recall', fontsize=12)
    plt.ylabel('Precision', fontsize=12)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    
    # Add textbox with metrics
    textstr = f'Average Precision: {ap:.4f}\nOptimal F1-Score: {f1_scores[optimal_idx]:.4f}'
    props = dict(boxstyle='round', facecolor='white', alpha=0.8)
    plt.text(0.05, 0.05, textstr, transform=plt.gca().transAxes, verticalalignment='bottom',
            bbox=props, fontsize=12)
    
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'precision_recall_curve_dim_{bottleneck_dim}.png', dpi=300)
    plt.show()
    
    return optimal_threshold, ap

# Function to plot ROC curves for comparison
def plot_roc_curves(results_dict):
    plt.figure(figsize=(10, 8))
    
    for dim, result in results_dict.items():
        plt.plot(result['fpr'], result['tpr'], linewidth=2,
                 label=f'Bottleneck Dim={dim}, AUC={result["auc"]:.4f}')
    
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title('ROC Curves for Different Bottleneck Dimensions', fontsize=14)
    plt.legend(loc="lower right")
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.savefig('roc_curves.png', dpi=300)
    plt.show()

# Train models with different bottleneck dimensions and evaluate
results = {}
optimal_thresholds = {}
reconstruction_errors_by_dim = {}
labels_by_dim = {}

for bottleneck_dim in bottleneck_dimensions:
    print(f"\n{'='*80}")
    print(f"Training model with bottleneck dimension: {bottleneck_dim}")
    print(f"{'='*80}")
    
    # Initialize and train the model
    model = Autoencoder(bottleneck_dim).to(device)
    model, losses = train_model(model, normal_train_loader, epochs, learning_rate)
    
    # Compute reconstruction errors on test set
    reconstruction_errors, labels = compute_reconstruction_error(model, test_loader)
    reconstruction_errors_by_dim[bottleneck_dim] = reconstruction_errors
    labels_by_dim[bottleneck_dim] = labels
    
    # Find optimal threshold using precision-recall curve
    optimal_threshold, ap = plot_precision_recall_curve(reconstruction_errors, labels, bottleneck_dim)
    optimal_thresholds[bottleneck_dim] = optimal_threshold
    
    # Evaluate model performance
    performance = evaluate_performance(reconstruction_errors, labels, optimal_threshold)
    results[bottleneck_dim] = performance
    
    # Plot histogram for all dimensions
    accuracy = plot_error_histogram(reconstruction_errors, labels, optimal_threshold, bottleneck_dim)
    
    # Plot heatmap of reconstruction errors by digit
    plot_error_heatmap_by_digit(reconstruction_errors, labels, bottleneck_dim)
    
    # Visualize some reconstructions
    recon_fig = visualize_reconstruction(model, test_loader)
    recon_fig.savefig(f'reconstructions_dim_{bottleneck_dim}.png', dpi=300)
    plt.close(recon_fig)
    
    # Print detailed results
    print("\nPerformance Metrics for Bottleneck Dimension =", bottleneck_dim)
    print(f"Optimal Threshold: {optimal_threshold:.6f}")
    print(f"Precision: {performance['precision']:.4f}")
    print(f"Recall: {performance['recall']:.4f}")
    print(f"F1-Score: {performance['f1_score']:.4f}")
    print(f"AUC-ROC: {performance['auc']:.4f}")
    print(f"Accuracy: {accuracy:.4f}")

# Plot ROC curves for comparison
plot_roc_curves(results)

# Create a comparative visualization of all bottleneck dimensions
plt.figure(figsize=(12, 8))

metrics = ['auc', 'precision', 'recall', 'f1_score']
values = {m: [results[dim][m] for dim in bottleneck_dimensions] for m in metrics}

x = np.arange(len(bottleneck_dimensions))
width = 0.2
multiplier = 0

colors = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12']

for attribute, metric_values in values.items():
    offset = width * multiplier
    bars = plt.bar(x + offset, metric_values, width, label=attribute.upper())
    
    # Add values on top of bars
    for bar in bars:
        height = bar.get_height()
        plt.annotate(f'{height:.3f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
    
    multiplier += 1

plt.xlabel('Bottleneck Dimension', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.title('Performance Metrics Comparison Across Bottleneck Dimensions', fontsize=14)
plt.xticks(x + width * (len(metrics) - 1) / 2, bottleneck_dimensions)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=len(metrics))

plt.ylim(0, 1.05)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('metrics_comparison.png', dpi=300)
plt.show()

# Print comparison of all dimensions
print("\nComparison of Different Bottleneck Dimensions:")
print(f"{'Dimension':^10} | {'AUC':^8} | {'F1-Score':^8} | {'Precision':^10} | {'Recall':^8} | {'Threshold':^10}")
print(f"{'-'*10} | {'-'*8} | {'-'*8} | {'-'*10} | {'-'*8} | {'-'*10}")
for dim in bottleneck_dimensions:
    result = results[dim]
    print(f"{dim:^10} | {result['auc']:.4f} | {result['f1_score']:.8f} | {result['precision']:.8f} | {result['recall']:.8f} | {result['threshold']:.8f}")

# Save the best model
best_dim = max(results.keys(), key=lambda k: results[k]['auc'])
best_model = Autoencoder(best_dim).to(device)
best_model, _ = train_model(best_model, normal_train_loader, epochs, learning_rate)
torch.save(best_model.state_dict(), f'autoencoder_best_dim_{best_dim}.pth')
print(f"\nBest model saved with bottleneck dimension: {best_dim}")
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Function to plot confusion matrix
def plot_confusion_matrix(reconstruction_errors, labels, threshold, bottleneck_dim):
    binary_labels = (labels == normal_digit).astype(int)
    predictions = (reconstruction_errors <= -threshold).astype(int)
    
    cm = confusion_matrix(binary_labels, predictions)
    
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Anomaly', 'Normal'], 
                yticklabels=['Anomaly', 'Normal'])
    plt.title(f'Confusion Matrix (Bottleneck Dim: {bottleneck_dim})')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.savefig(f'confusion_matrix_dim_{bottleneck_dim}.png', dpi=300)
    plt.show()

# Generate confusion matrices for each dimension
for bottleneck_dim in bottleneck_dimensions:
    reconstruction_errors = reconstruction_errors_by_dim[bottleneck_dim]
    labels = labels_by_dim[bottleneck_dim]
    threshold = optimal_thresholds[bottleneck_dim]
    
    plot_confusion_matrix(reconstruction_errors, labels, threshold, bottleneck_dim)
```

### Key Observations:

1. **False Positive Reduction**: As the bottleneck dimension increases from 8 to 32, there's a significant reduction in false positives (429 → 311 → 265), which is particularly important for anomaly detection systems where false alarms can be costly.

2. **True Positive Performance**: The bottleneck dimension of 16 achieves the highest number of true positives (8834), while dimensions 8 and 32 have nearly identical true positive counts (8702 and 8703).

3. **Optimal Dimension Trade-offs**:
   - Dimension 8: Highest false positive rate (43.7%), suggesting insufficient capacity to properly model normal data
   - Dimension 16: Best balance between detecting anomalies (highest true positives) and limiting false positives
   - Dimension 32: Lowest false positive rate (27.0%) but slightly fewer true positives than dimension 16

4. **Normal Class Recognition**: As bottleneck dimension increases, the model gets progressively better at correctly identifying normal samples (553 → 671 → 717), which directly impacts the false positive rate.

5. **Anomaly Detection Sensitivity**: The dimension 16 model has the lowest miss rate for anomalies (only 184 false negatives compared to ~315 for the other models), suggesting optimal sensitivity for detecting unusual patterns.

These findings align with theoretical expectations in anomaly detection: too small a bottleneck (dim=8) restricts the model's ability to properly characterize normal data, leading to high false positive rates. Conversely, while a larger bottleneck (dim=32) improves modeling of normal data (reducing false positives), it might slightly reduce sensitivity to anomalies.

For most anomaly detection applications where false alarms are particularly problematic, the dimension 32 model would be preferable despite its marginally lower anomaly detection rate, as it achieves the best false positive performance. However, if missing anomalies is particularly costly, dimension 16 provides the best overall detection rate with reasonable false positive control.
